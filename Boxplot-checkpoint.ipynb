## Horizontal boxplot`

```python
sns.boxplot(x="tip", y="day", hue="smoker", data=tips1) #Note: if you want to draw the plot in horizontal then assign y to categorical value and x is to numerical and vice versa is default vertical.
```

## Dodge property 
- means that the bars in boxplot are merged in this propery.
- as smoker into days bars

```python
sns.boxplot(x="tip", y="day", hue="smoker", dodge=False, data=tips1)
```

## Palette property
- Rocket is color palette

```python
sns.boxplot(x="day", y="tip", hue="smoker", data=tips1, palette="rocket")
```

## Head(number) function and showmeans

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
kashti=sns.load_dataset("titanic")
kashti.head(50) # this function will "head(50)" only show dataset upto 50 rows
```

## showmeans property, meanprops with marker, markersize, markeredgecolor.

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
tips=sns.load_dataset("tips")
tips
```

```python
sns.boxplot(x="sex",
            y="total_bill",
            meanprops={
                "marker":"*",
                "markersize":"9",    #meanprops property shows means in the bars of boxplot with marker,markersize,markeredgecolor.
                "markeredgecolor":"yellow"
                },
            showmeans=True,
            data=tips)
plt.title("Boxplot with means properties {marker,markersize,markeredgecolor}", size=16, weight="bold")
plt.xlabel("Males and Females")
plt.ylabel("Total bill")
```
