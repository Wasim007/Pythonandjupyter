```python
import pandas as pd                 #this commands are used to import userdefined dataset into jupyter and to visualize the dataset
gdpdata=pd.read_csv("GDP.csv")
gdpdata
```

## (title of figure) as mention below 

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30)) #by this command the size of figure is increased
sns.lineplot(x="Year", y="Afghanistan" , data=gdpdata)
plt.title("Afghanistan ka GDP")
plt.show()
```
